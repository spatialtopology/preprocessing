```python
sbref_dir = '/Users/h/Documents/projects_local/sandbox/fmriprep_bold/sbref'
flist = glob.glob(join(sbref_dir, '*sbref.nii'))
for fname in flist:
    plotting.plot_stat_map(image.load_img(fname) , black_bg=True, draw_cross=False, cut_coords=(0,0,0))
    plt.savefig(join(sbref_dir,  os.path.basename(fname[:-4] )+'.png'))
    
```
